```python
#find the parameters that can define a multivariate lineal regression for this data.

#make each variable a numerical list to input to the model.
with open('insurance.csv') as insurance_csv:
    reader = csv.DictReader(insurance_csv, delimiter = ',')
    age_v = []
    sex_v = []
    bmi_v = []
    children_v = []
    smoker_v = []
    region_v = []
    charges_v = []
    for item in reader:
        age_v.append(float(item['age']))
        bmi_v.append(float(item['bmi']))
        charges_v.append(float(item['charges']))
        children_v.append(float(item['children']))
        if item['sex'] == 'female':
            sex_v.append(0)
        elif item['sex'] == 'male':
            sex_v.append(1)
        else:
            pass
        if item['smoker'] == 'no':
            smoker_v.append(0)
        elif item['smoker'] == 'yes':
            smoker_v.append(1)
        else:
            pass
        if item['region'] == 'northeast':
            region_v.append(1)
        elif item['region'] == 'northwest':
            region_v.append(2)
        elif item['region'] == 'southeast':
            region_v.append(3)
        else:
            region_v.append(4)

#printing to see if it worked.
#print(age_v)
#print(sex_v)
#print(bmi_v)
#print(children_v)
#print(smoker_v)
#print(region_v)
#print(charges_v)

#import pandas, numpy and statsmodels.
import pandas as pd
import statsmodels.api as sm
import numpy as np

#define a DataFrame with pd
data = {
    'age': age_v,
    'sex': sex_v,
    'bmi': bmi_v,
    'children': children_v,
    'smoker': smoker_v,
    'region':region_v,
    'charges':charges_v
}
df = pd.DataFrame(data)

#create separate columns for each category of region (either , 0 or 2, 0 or 3...). Dropping the first accounts for multicollinearity (if r2,r3,r4 = 0
#then the individual is inherently in r1
df = pd.get_dummies(df, columns=['region'], drop_first = True)

#convert booleans to int
df = df.astype({col: int for col in df.select_dtypes('bool').columns})

#define dependent and independent vars
X = df.drop('charges', axis = 1) #independent
y = df['charges'] #dependent

#set intercept
X['constant'] = 1.0
X = X.apply(pd.to_numeric, errors = 'coerce')

#check for null values
print(df.isnull().sum())

#assure that X and y are arrays.
X_np = np.asarray(X)
y_np = np.asarray(y)

#check datatypes
print(X_np.dtype, y_np.dtype)
print(X.dtypes)

#print descriptive box
print(df.describe(include='all'))

#print each array (one has all dependent variables (matrix) and one is the cost vector) 
print(X_np[:5])
print(y_np[:5])

#Apply regression
model = sm.OLS(y_np,X_np).fit()

#Summary statistics
print(model.summary())
```
